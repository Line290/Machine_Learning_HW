```python
import os
import sys
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, Bidirectional, LSTM, GRU
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
DATA_PATH = os.path.join('data', 'ShanghaiCompositeIndex.csv')
df = pd.read_csv(DATA_PATH, encoding='gbk')
df.head()
```

```python
df = df.drop(['股票代码'.decode("utf-8"), '名称'.decode("utf-8")], axis=1)
df.head()
```

```python
def plot_series(values1, values2, xlabel=None, ylabel=None, color=['b', 'r'], legend=None):
    xx = np.arange(1, len(values1) + 1, 1)
    plt.figure(figsize=(200,100))
    plt.plot(xx, values1, color[0], label=legend[0])
    plt.plot(xx, values2, color[1], label=legend[1])
    plt.legend(loc = 'upper left')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
plot_series(df['最高价'.decode('utf-8')].values[::-1], 
            df['最低价'.decode('utf-8')].values[::-1],
            xlabel='Days', 
            ylabel='The MAX and MIN price on every day', 
            color=['b', 'r'], 
            legend=['max price', 'min price'])
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
X = df.values[:-2,2][::-1].reshape(-1,1).astype(np.float32)
print X.shape
X = scaler.fit_transform(X)
print X
```

```python
train, test = X[:-5], X[-5:]
print train.shape, test.shape
```

```python
# train, test = train_test_split(X, test_size=0.3, shuffle=False)
# print train.shape
# print test.shape
```

```python
X_train, y_train = train[:, np.newaxis, :], X[1:-4]
X_test, y_test = X[-6:-1][:, np.newaxis, :], test
print X_train.shape, y_train.shape
print X_test.shape, y_test.shape
```

```python
EPOCHS = 50
TEST_SIZE = 0.3
TIME_AHEAD = 1 #prediction step
BATCH_SIZE = 1
UNITS = 100
```

```python
def create_symbol(model_name='BiLSTM', units=10, activation='linear', time_ahead=1):
    model = Sequential()
    if model_name == 'LSTM':
        model.add(LSTM(units, input_shape=(1, time_ahead)))
    elif model_name == 'BiLSTM':
        model.add(Bidirectional(LSTM(units), input_shape=(1, time_ahead)))
    elif model_name == 'GRU':
        model.add(GRU(units, input_shape=(1, time_ahead)))
    else:
        raise ValueError("Wrong model name")
    model.add(Dense(1))
    model.add(Activation('linear'))
    return model
```

```python
model = create_symbol(model_name='BiLSTM', units=UNITS, time_ahead=TIME_AHEAD)
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2)
```

```python
p_test = []
p_t = X_test[0]
for i in range(5):
    p_t = model.predict(p_t[np.newaxis,:,:])
    p_test.append(float(p_t))
```

```python
p_test_inv = scaler.inverse_transform(np.asarray(p_test).reshape(-1,1))
score = np.sqrt(mean_squared_error(y_test_inv, p_test_inv[:y_test_inv.shape[0]]))
print 'Test RMSE: %.2f' % (score)
```

```python
def plot_series_prediction(true_values, train_predict, test_predict, time_ahead=1, title=None,
                           xlabel=None, ylabel=None, color=['green','red','blue'], legend=[None,None,None]):    
    TOOLS = 'pan,wheel_zoom,box_zoom,reset,save,box_select'
    #x axis
    xx = np.array(range(true_values.shape[0]))
    xx1 = np.array(range(time_ahead,len(train_predict)+time_ahead))
    xx2 = np.array(range(len(train_predict),len(true_values)))
    #figure
#     p = plt.figure(title=title, tools=TOOLS)
    plt.figure(figsize=(20,10))
#     plt.plot(xx, true_values.squeeze(), label=legend[0], color=color[0])
#     plt.plot(xx1, train_predict.squeeze(), label=legend[1], color=color[1])    
    plt.plot(xx2, test_predict.squeeze(), label=legend[2], color=color[2])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc = "upper left")
    plt.show()
```

```python
pred_train = model.predict(X_train)
pred_train = scaler.inverse_transform(pred_train)
print X_inv.shape, pred_train.shape, pred_test.shape
plot_series_prediction(X_inv, pred_train, p_test_inv[:y_test_inv.shape[0]], time_ahead=TIME_AHEAD,
                       title='LSTM prediction', xlabel='Days', ylabel='Value of ', 
                       legend=['True value','Training set','Test prediction'])
```
